# Simple example of dependecy parser used for noun phrases detection

```python
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
```

```python
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
```

```python
s = "The restaurant service was very good, but I'm not sure about the food quality"
```

```python
doc = nlp(s)
components = []
for chunk in doc.noun_chunks:
    components.append({
        'text': chunk.text, 'root': chunk.root.text, 'root_dep': chunk.root.dep_, 
        'root_head': chunk.root.head.text
    })
pd.DataFrame(components)
```

```python
displacy.render(doc)
```
