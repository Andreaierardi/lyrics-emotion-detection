# Explore a dataset of argumentative and dialog texts
> Christian Stab and Iryna Gurevych (2014) Annotating Argument Components and Relations in Persuasive Essays. In: Proceedings of the the 25th International Conference on Computational Linguistics (COLING 2014), p.1501-1510, Ireland, Dublin.

```python
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
```

```python
import spacy
import os
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
dataset_folder = 'data/brat-project/'
docs = []
for f in tqdm(os.listdir(dataset_folder)):
    if f.endswith('.txt'):
        with open(os.path.join(dataset_folder, f), 'r') as data:
            text = data.read()
            docs += [s for s in nlp(text).sents]
```

## Surveys

```python
c = 'but'
for z, doc in enumerate(docs):
    i = [j for j, w in enumerate(doc) if w.lemma_ == c]
    if len(i) > 0:
        pos = i[0]
        if len(doc[:pos]) > 0:
            print(doc[:pos])
            print(c)
            print(doc[pos+1:], '\n')
        else:
            print(docs[z-1])
            print(c)
            print(doc[pos+1:], '\n')
```
